```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MWData.zip", delimiter = "\t")
df.columns = ['id','event','device','channel','code','size','data']
df.head()
```

```python
eeg = df.iloc[0:20000, :-1]
xyz = df.iloc[0:20000,-1]
signals = xyz.str.split(',', expand=True).astype(float)
```

```python
epoch = signals.iloc[:,400:900] #400 and 900 - stimulus times
epoch.fillna(0)
```

```python
min_arr = epoch.min(axis = 1)
max_arr = epoch.max(axis = 1)

large = max_arr-min_arr
```

```python
epochnew = epoch
for i in range(500):
  epochnew[i+400] = epochnew[i+400]-min_arr
  epochnew[i+400] = epochnew[i+400]/large
```
